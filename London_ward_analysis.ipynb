```python
import pandas as pd, numpy as np, seaborn as sns, matplotlib.pyplot as plt, geopandas as gpd, folium
```

Reading in the shapefiles and ward data

```python
df0 = gpd.read_file('statistical-gis-boundaries-london/ESRI/London_Ward_CityMerged.shp')

df1 = pd.read_csv("ward_data.csv", encoding = 'ISO-8859-1')

df1 = df1.iloc[:,[0,2,3,8,10,13,15,-5,-8, 30, 29]]
```

```python
df1.head(1)
```

```python
df0.head(1)
```

```python
df2 = df0.merge(df1, how = 'left', left_on = 'GSS_CODE', right_on = 'New code')
```

```python
df2.columns.values
```

```python
plt1 = df2.plot(column = 'Mean Age - 2013', cmap = 'Purples', scheme = 'quantiles')
plt.style.use('seaborn-white')
plt1.axes.xaxis.set_ticklabels([])
plt1.axes.yaxis.set_ticklabels([])
plt.figure(figsize = (15, 15))
plt.show()
```
